Now, to explore how the houses are distributed over the states and the property types.

```python
#Count by state

count_by_state= (
 df.groupby("state")
  .count()
  .sort_values("town", ascending= False)["price"]
   )

# BarChart for House counts per state 
count_by_state.plot(
   kind = "bar",
   xlabel = "State",
   ylabel = "Counts",
   title= "Distribution of Houses per State "

);
```

```python

#Count by title

count_by_title= (
    df.groupby("title")
   .count()["town"]
   .sort_values(ascending=False)
)

# BarChart for House counts by title 
count_by_title.plot(
   kind = "bar",
   xlabel = "House Types",
   ylabel = "Counts",
   title= "Distribution of Houses per Title ",
   color = "green"

);
```

From the two BarChart it can be seen that majority of the houses in our dataset are from "Lagos", and most of the property types are "Detached Duplex" .

Thus, we'll be focusing on Detached Duplex in Lagos to build our Machine learning model, because with that we'll have more observations to deal with.
